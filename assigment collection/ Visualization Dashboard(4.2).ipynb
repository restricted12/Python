```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
# creating data 
import pandas as pd
visualizations={
    "months" :['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    "salaries" : [85792, 93402, 67000, 120381, 102380, 57000, 85000, 93020, 120943, 104982, 85000, 127000],
    "employee_names" : ['John Doe', 'Alice Smith', 'Bob Johnson', 'Carol White', 'David Brown', 
                  'Eve Black', 'Frank Green', 'Grace Wilson', 'Hank Lewis', 'Ivy Young', 
                  'Jake King', 'Lily Adams'],
    "positions" : ['Developer', 'Manager', 'Analyst', 'Engineer', 'Team Lead', 
             'Designer', 'QA Specialist', 'IT Support', 'Consultant', 
             'Data Scientist', 'DevOps', 'Product Owner'],
    "ages" :[34, 45, 28, 51, 33, 39, 41, 47, 29, 55, 32, 50],
    "num_employees" :[12, 18, 23, 30, 22, 40, 25, 35, 10, 15, 42, 11]
}
# displaying data
xe=pd.DataFrame(visualizations)
print(xe)
```

```python
# converting the data in to csv file
xe.to_csv("C:/Users/ezepayooner/OneDrive/data_visualizations.csv", index=False)
# reading(displaying) the data 
xe_read = pd.read_csv("C:/Users/ezepayooner/OneDrive/data_visualizations.csv")
print(xe_read)
```

```python
#  A line plot showing the trend of sales over the past year
import matplotlib.pyplot as plt

x = visualizations["months"]
y = visualizations["salaries"]
plt.figure(figsize=(10, 5))
plt.plot(x, y,marker="o",c="r",ls="--")
plt.title("TREND OF SALARIES OVER THE PAST YEAR 2023")
plt.xlabel("MONTHS")
plt.ylabel("SALARIES")
plt.grid(True)
plt.show()
```

```python
#  A bar plot comparing the number of employees in different departments.

import matplotlib.pyplot as plt

# Assuming visualizations is a dictionary with your data
x = visualizations["positions"]
y = visualizations["num_employees"]

# Create a new figure with a specific size
plt.figure(figsize=(15, 5))

# Set the title and labels
plt.title("Number of Employees in Different Departments".upper())
plt.xlabel("Departments".upper())
plt.ylabel("Number of Employees".upper())

# Create a bar chart
plt.bar(x, y)

# Add gridlines
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Show the plot
plt.show()
```

```python
# A scatter plot showing the relationship between employee age and salary.
import seaborn as sns
import matplotlib.pyplot as plt

x = visualizations["ages"]
y = visualizations["salaries"]
plt.figure(figsize=(10, 5))
# Using seaborn's lineplot with x and y as keyword arguments
sns.scatterplot(x=visualizations["ages"],y=visualizations["salaries"], c="r")

# Optional: Add labels and title for clarity
plt.title("employee age vs salary".upper())
plt.xlabel("Ages".upper())
plt.ylabel("Salaries".upper())
plt.grid(True)
plt.show()
```
